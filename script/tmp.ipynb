```python
import pandas as pd

def make_corr(x, y, x_label, y_label):
    df_x = pd.DataFrame(x, columns=[x_label])
    df_y = pd.DataFrame(y, columns=[y_label])

    df = pd.concat([df_x, df_y], axis=1)

    corr = df.corr()
    return corr
```

```python
corr_mb1 = make_corr(mean1, ratioB1, "L1 mean", "L1 ratio_bonded")
corr_mb2 = make_corr(mean2, ratioB2, "L2 mean", "L2 ratio_bonded")
corr_mb3 = make_corr(mean3, ratioB3, "L3 mean", "L3 ratio_bonded")

print(corr_mb1)
print()
print(corr_mb2)
print()
print(corr_mb3)
```

```python
corr_md1 = make_corr(mean1, dev1, "L1 mean", "L1 deviation")
corr_md2 = make_corr(mean2, dev2, "L2 mean", "L2 deviation")
corr_md3 = make_corr(mean3, dev3, "L3 mean", "L3 deviation")

print(corr_md1)
print()
print(corr_md2)
print()
print(corr_md3)
```

```python
corr_bd1 = make_corr(ratioB1, dev1, "L1 ratio_bonded", "L1 deviation")
corr_bd2 = make_corr(ratioB2, dev2, "L2 ratio_bonded", "L2 deviation")
corr_bd3 = make_corr(ratioB3, dev3, "L3 ratio_bonded", "L3 deviation")

print(corr_bd1)
print()
print(corr_bd2)
print()
print(corr_bd3)
```
