```python
new_dict1 = {}
new_dict2 = {}
new_dict3 = {}
with open("../data/dic/L1_data_1.pkl", "rb") as tf:
    new_dict1 = pickle.load(tf)
with open("../data/dic/L2_data_1.pkl", "rb") as tf:
    new_dict2 = pickle.load(tf)
with open("../data/dic/L3_data_1.pkl", "rb") as tf:
    new_dict3 = pickle.load(tf)
```

```python
mean1 = []
dev1 = []
ratioB1 = []
for key in new_dict1:
    ratioB1.append(new_dict1[key]["ratio_of_volume"])
    mean1.append(new_dict1[key]["mean_volume"])
    dev1.append(new_dict1[key]["deviation_of_volume"])
```

```python
mean2 = []
dev2 = []
ratioB2 = []
for key in new_dict2:
    ratioB2.append(new_dict2[key]["ratio_of_volume"])
    mean2.append(new_dict2[key]["mean_volume"])
    dev2.append(new_dict2[key]["deviation_of_volume"])
```

```python
mean3 = []
dev3 = []
ratioB3 = []
for key in new_dict3:
    ratioB3.append(new_dict3[key]["ratio_of_volume"])
    mean3.append(new_dict3[key]["mean_volume"])
    dev3.append(new_dict3[key]["deviation_of_volume"])
```

```python
plt.scatter(mean1, ratioB1)
plt.title("L1 relation between mean volume and ratio of bonded")
plt.xlabel("mean volume")
plt.ylabel("ratio of bonded")
plt.grid(True)
```

```python
plt.scatter(mean2, ratioB2)
plt.title("L2 relation between mean volume and ratio of bonded")
plt.xlabel("mean volume")
plt.ylabel("ratio of bonded")
plt.grid(True)
```

```python
plt.scatter(mean3, ratioB3)
plt.title("L3 relation between mean volume and ratio of bonded")
plt.xlabel("mean volume")
plt.ylabel("ratio of bonded")
plt.grid(True)
```
